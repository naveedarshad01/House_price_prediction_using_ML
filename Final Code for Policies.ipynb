```python
import os
import zipfile
import shutil
import glob
from datetime import datetime

# Step 1: Unzip files and remove zipped files
source_folder = r"C:\Digitata Work\WEEK 33\16082023\Files\RAW FILES"
unzip_folder = os.path.join(source_folder, "Unzipped")
os.makedirs(unzip_folder, exist_ok=True)

for zip_file in os.listdir(source_folder):
    if zip_file.endswith(".zip"):
        with zipfile.ZipFile(os.path.join(source_folder, zip_file), 'r') as zip_ref:
            zip_ref.extractall(unzip_folder)
        os.remove(os.path.join(source_folder, zip_file))

# Step 2: Clear contents of Policy CSV folder and move Policy CSV files
policy_csv_folder = os.path.join(source_folder, "Policy CSV")
if os.path.exists(policy_csv_folder):
    shutil.rmtree(policy_csv_folder)
os.makedirs(policy_csv_folder)

for csv_file in glob.glob(os.path.join(unzip_folder, "*.csv")):
    if "Policy" in csv_file and "parameters" not in csv_file:
        shutil.move(csv_file, os.path.join(policy_csv_folder, os.path.basename(csv_file)))

# Step 3: Clear contents of Parameters CSV folder and move Parameters CSV files
parameters_csv_folder = os.path.join(source_folder, "Parameters CSV")
if os.path.exists(parameters_csv_folder):
    shutil.rmtree(parameters_csv_folder)
os.makedirs(parameters_csv_folder)

for csv_file in glob.glob(os.path.join(unzip_folder, "*.csv")):
    if "Policy" in csv_file and "parameters" in csv_file:
        shutil.move(csv_file, os.path.join(parameters_csv_folder, os.path.basename(csv_file)))

# Step 4: Combine Policy CSV files
policy_output_file = os.path.join(policy_csv_folder, f"Policy_File_{datetime.now().strftime('%Y%m%d_%H%M%S')}.csv")
policy_files = glob.glob(os.path.join(policy_csv_folder, "*.csv"))
with open(policy_output_file, 'w') as output:
    for i, file in enumerate(policy_files):
        with open(file, 'r') as f:
            if i == 0:  # Include header from the first file
                output.write(f.read())
            else:
                lines = f.readlines()[1:]  # Skip header for other files
                output.writelines(lines)

# Step 5: Combine Parameters CSV files
parameters_output_file = os.path.join(parameters_csv_folder, f"Parameters_File_{datetime.now().strftime('%Y%m%d_%H%M%S')}.csv")
parameters_files = glob.glob(os.path.join(parameters_csv_folder, "*.csv"))
with open(parameters_output_file, 'w') as output:
    for i, file in enumerate(parameters_files):
        with open(file, 'r') as f:
            if i == 0:  # Include header from the first file
                output.write(f.read())
            else:
                lines = f.readlines()[1:]  # Skip header for other files
                output.writelines(lines)
```
